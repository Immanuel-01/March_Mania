```python
!pip uninstall streamlit_jupyter -y
!pip install --upgrade streamlit_jupyter

import streamlit as st
from streamlit_jupyter import StreamlitPatcher

# Patch Streamlit to run inside Jupyter
StreamlitPatcher().jupyter()

# Streamlit UI
st.title("🏀 NCAA Match Predictor")
st.write("Select two teams to predict the winner.")

# Example UI Elements
team1 = st.text_input("Enter Team 1")
team2 = st.text_input("Enter Team 2")

if st.button("Predict"):
    st.success(f"🏆 Predicted Winner: {team1 if team1 > team2 else team2}")
    st.write("🔥 Win Probability: 65%")
```
